Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

    A probability distribution is a function that describes the likelihood of obtaining different possible outcomes in a random event. Although the values of the event are meant to be random, the probability distribution predicts the likelihood of each value occurring.

Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

     True random numbers are generated by natural processes, whereas pseudo-random numbers are generated by a deterministic algorithm. While true random numbers are ideal, they are difficult to generate and often require expensive hardware. Pseudo-random numbers are generated by mathematical algorithms that can produce sequences that approximate true randomness. These sequences are considered "good enough" for many applications.

Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

    The two main factors that influence the behaviour of a "normal" probability distribution are the mean and standard deviation. The mean represents the central tendency of the distribution, while the standard deviation represents the dispersion or spread of the distribution.

Q4. Provide a real-life example of a normal distribution.

     A real-life example of a normal distribution is the distribution of heights or weights of a population. In general, most people will fall close to the mean height or weight, with fewer individuals being significantly taller or shorter or heavier or lighter.

Q8. What is the relationship between exponentiation and logarithms?

    Exponentiation and logarithms are inverse operations. If x^y = z, then log_x(z) = y. In other words, taking the logarithm of a number tells you what exponent is required to produce that number, and raising a number to an exponent tells you what number is produced by that exponent.



Q9. What are the three logarithmic functions that Python supports?

    The three logarithmic functions that Python supports are the natural logarithm (math.log), the base-10 logarithm (math.log10), and the base-2 logarithm (math.log2).
