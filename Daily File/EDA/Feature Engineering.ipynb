```python
col_with_nan=[col for col in df.columns if df[col].isna().sum()>1 ]
```

```python
for col in col_with_nan:
    print(col," --> Having Missing Value of ",np.round((df[col].isnull().mean())*100,4),"%")
```

```python
categorical_features=[col for col in df.columns if df[col].dtypes == 'object']
```

```python
data=df.copy()
#Replace with "Missing"
def replace_cate_colm_nan(data,cat_fetures):
    data[cat_fetures]=data[cat_fetures].fillna('Missing')
    return data


replace_cate_colm_nan(data,categorical_features)
```

```python
data=df.copy()
#Replace with "Mode"
def replace_cate_colm_nan(data,cat_fetures):
    
    for i in cat_fetures:
        data[i]=data[i].fillna(data[i].mode()[0])
    return data

replace_cate_colm_nan(data,categorical_features)
```

```python
numerical_features=[col for col in df.columns if df[col].dtypes != 'object']
```

```python
numerical_features_with_nan=[col for col in df.columns if df[col].dtypes != 'object' and df[col].isna().sum()>=1]
```

```python
numerical_features_with_nan
```

```python
for col in numerical_features_with_nan:
    
    df[col+' nan']=np.where(df[col].isna(),1,0)
    df[col]=df[col].fillna(df[col].median())
```
