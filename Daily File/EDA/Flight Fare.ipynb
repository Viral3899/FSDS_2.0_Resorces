```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Data_Train.xlsx')
```

```python
df
```

```python
df.info()
```

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
df
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_time_only']=df['Arrival_Time'].str.split(" ").str[0]
df['Dep_time_only']=df['Dep_Time'].str.split(" ").str[0]
```

```python
df
```

```python
df['Arrival_hr']=df['Arrival_time_only'].str.split(':').str[0].astype(int)
df['Arrival_min']=df['Arrival_time_only'].str.split(':').str[1].astype(int)

df['Dep_hr']=df['Dep_time_only'].str.split(':').str[0].astype(int)
df['Dep_min']=df['Dep_time_only'].str.split(':').str[1].astype(int)
```

```python
df
```

```python
df.drop(['Dep_Time','Arrival_Time'],axis=1,inplace=True)
```

```python
df
```

```python
duration=list(df['Duration'])
for i in range(len(duration)):
    if 'm' not in duration[i]:
        duration[i]=duration[i]+" 0m"
    elif 'h' not in duration[i]:
        duration[i]="0h "+duration[i]
        
        
        
df['Duration']=pd.DataFrame(duration)
df['Duration_hr']=df['Duration'].str.split(" ").str[0]
df['Duration_min']=df['Duration'].str.split(" ").str[1]
df[['Duration_hr','Duration_min']]=df[['Duration_hr','Duration_min']].replace('[A-Za-z]','',regex=True)
```
