```python
df=df.sort_values(by='Product Vendor')
```

```python
from datetime import datetime, timedelta

# Create a dictionary that maps vendors to lists of order timestamps
order_timestamps = {}
for index,order in df.iterrows():
    vendor = order['Customer Email']
    timestamp = order['Order Date']
    if vendor not in order_timestamps:
        order_timestamps[vendor] = []
#         print(order_timestamps)
    order_timestamps[vendor].append(timestamp)
# print(order_timestamps.items())
# Calculate the average gap between orders for each vendor
avg_gaps = {}
for vendor, timestamps in order_timestamps.items():
    timestamps=sorted(timestamps)
    print(vendor,'has no of order date in data -->>',len(timestamps))
    total_gap = timedelta(0)
    for i in range(1, len(timestamps)):
        gap = timestamps[i] - timestamps[i-1]
        total_gap += gap
        
    if len(timestamps)==1:
        print(f'there is only 1 order for vendor name {vendor}---> Not Possible to calcluate gap')
        continue
    else:
        avg_gap = total_gap / (len(timestamps) - 1)
        avg_gaps[vendor] = avg_gap

# Print out the results
for vendor, avg_gap in avg_gaps.items():
    print(f'Average gap between orders for  <----> {vendor} <--->  :  <----> {avg_gap}')
```
